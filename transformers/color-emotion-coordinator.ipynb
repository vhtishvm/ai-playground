```python
import numpy as np
```

```python
#Raw data
senA = "The rose is red"
senB = "The signal is red"


```

```python
#Assigning a single token for each word
np.random.seed(42)

encoded = {}

def encoding(sentence):
    
    vocab = sentence.split()
    for word in vocab:
        random_number = np.random.rand(1,2)
        if word not in encoded.keys():
            encoded[word] = random_number
    return encoded

encodedA = encoding(senA)
encodedB = encoding(senB)

tokensA = [encodedA[word] for word in senA.split()]

#print(tokensA)

tokensB = [encodedB[word] for word in senB.split()]
#print(tokensB)


#Making vertical matrices
X_A = np.vstack(tokensA)
X_B = np.vstack(tokensB)

print(f"The first matrix is: {X_A}")
print(f"The second matrix is: {X_B}")
```
