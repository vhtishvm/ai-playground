```python
import numpy
import tiktoken
tokenizer = tiktoken.get_encoding("cl100k_base")
print(f"Vocabulary size: {tokenizer.n_vocab:,} tokens")
```

```python
text = "My name is ahtisham: "
vocab = tokenizer.n_vocab
print(vocab)
tokens = tokenizer.encode(text)
tokens
```
